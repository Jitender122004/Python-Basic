2) What is an interpreter in Python?
  -> An interpreter in Python is a program that reads and executes Python code line by line. It translates the high-level Python instructions into machine code that the computer can understand. Unlike compiled languages, where the entire code is translated at once, Python’s interpreter runs each line immediately, making it easier to test and debug. Popular Python interpreters include CPython, PyPy, and Jython.

  -> Pre-defined keywords in Python are reserved words that have special meanings and purposes in the language. They are used to define the syntax and structure of Python programs and cannot be used as variable names.

4) Can keyword be used as variable names?
  ->No, keywords cannot be used as variable names in Python.
Keywords are reserved words that have special meaning in the language (like if, while, class, def, etc.). Using them as variable names will result in a syntax error.
Example:
python
if = 5  # ❌ This will cause a syntax error
If you try this, Python will give an error because if is a keyword used for conditional statements. Always use non-keyword names for variables.

5) What is mutability in Python?
  -> Mutability refers to whether an object’s value can be changed after it is created.
Mutable objects: Can be modified after creation (e.g., change elements, add, remove).
Examples: list, dict, set

Immutable objects: Cannot be changed after creation. Any modification creates a new object.
Examples: int, float, str, tuple

Example:
python
# Mutable
a = [1, 2, 3]
a[0] = 100  # Allowed

# Immutable
b = (1, 2, 3)
# b[0] = 100  # ❌ Error – tuples are immutable
Mutability affects how variables behave in memory and during function calls.

6) Why are lists mutable, but tuples are immutable?
  ->Lists are mutable, meaning you can change, add, or remove elements after creation.
  Tuples are immutable, meaning once created, their elements cannot be changed.
  This design choice allows lists for dynamic data, and tuples for fixed or constant data that should not be altered.

7) What is the difference between == and is operators in Python?
 -> == checks value equality (if two variables have the same value).
   is checks identity (if two variables point to the same object in memory).

python
a = [1, 2]; b = [1, 2]
a == b  # True (values are same)
a is b  # False (different objects)

python

10) What is the difference between implicit and explicit type casting?
  -> Implicit casting: Python automatically converts types during operations
python

Explicit casting: You manually convert types using functions like int(), str(), float()

python

11) What is the purpose of conditional statements in Python?
Conditional statements allow programs to make decisions based on conditions. They control which block of code gets executed (e.g., using if, elif, else).

12) How does the elif statement work?
  -> elif (else if) lets you check multiple conditions after an if. It runs only if previous conditions are false.

python

13) What is the difference between for and while loops?
  ->for loop is used when you know how many times to loop (used with sequences).

while loop is used when you don’t know how many times to loop; it runs as long as a condition is true.

14) Describe a scenario where a while loop is more suitable than a for loop.
 -> A while loop is better when looping depends on a condition, not a fixed count.
Example: Asking for user input until they type "exit":

python
